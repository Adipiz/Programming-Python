### Exercise 1

1. Create sin(x) and cos(x) in 2 different plots

```python
import matplotlib.pyplot as plt
import numpy as np


x_i = 0
x_f = 2*np.pi
increment = 0.05

x = np.arange(x_i, x_f, increment)
y = np.cos(x)
z = np.sin(x)

plt.subplot(2,1,1)
plt.plot(x,y, 'b')
plt.title('Cosin function')
plt.ylabel('cos(x)')
plt.xlabel('x')
plt.grid()
plt.show()


plt.subplot(2,1,2)
plt.plot(x,z, 'r')
plt.title('Sin function')
plt.ylabel('sin(x)')
plt.xlabel('x')
plt.grid()
plt.show()
```
